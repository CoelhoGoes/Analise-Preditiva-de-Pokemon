## 2. Análise de Valores Nulos e Duplicados

### Contagem de valores nulos por coluna

```python
print(df.isnull().sum())
print(f"\n\nTotal de valores nulos no dataset: {df.isnull().sum().sum()}")
```

### Percentual de valores nulos por coluna

```python
print((df.isnull().sum() / len(df) * 100).round(2))
```

### Verificar registros duplicados

```python
print(f"Número de registros duplicados: {df.duplicated().sum()}")
```

### Separar colunas por tipo (numéricas e categóricas)

```python
print("Colunas numéricas:")
print(df.select_dtypes(include=['int64', 'float64']).columns.tolist())
print("\nColunas categóricas/texto:")
print(df.select_dtypes(include=['object']).columns.tolist())
```

## 5. Valores Únicos e Cardinalidade

### Número de valores únicos por coluna

```python
print(df.nunique())
```

### Valores únicos das colunas categóricas principais

```python
colunas_categoricas = df.select_dtypes(include=['object']).columns

for coluna in colunas_categoricas:
    print(f"\n{coluna}:")
    print(f"Valores únicos: {df[coluna].nunique()}")
    if df[coluna].nunique() <= 20:  # Mostrar apenas se tiver até 20 valores únicos
        print(f"Valores: {df[coluna].unique()}")
```
